```python
import logging
logging.basicConfig(filename = "Error.log",level = logging.DEBUG,format='%(asctime)s %(levelname)s %(message)s') 
def find_replaceSpecificText(specfic_text,replace_text,fileNm):
    try:
        with open(fileNm, 'r') as file:
            file_conts = file.read() # reading content from file
            remove_cont = file_conts.replace(specfic_text,replace_text)
            file.close()
            logging.info("executed succesfully")
    except Exception as e:
        logging.error("Error has happended")
        logging.exception(" Exception has happended " + str(e))
    try:
        with open(fileNm, 'w') as file:
            file.write(file_conts)
            file.close()
            logging.info("executed succesfully")
    except Exception as e:
        logging.error("Error has happended")
        logging.exception(" Exception has happended " + str(e))
```

` Q2.Demonstrate use of abstract class, multiple inheritance and decorator in python using examples.`

```python
"""
 Python Program to demonstrate multiple inheritance
"""
# Parent Class-1
class EmpInfo(object):
    def __init__(self,name,eid):
        self.name = name
        self.eid = eid

# Parent Class-2
class OtherDtls(object):
    def __init__(self,salary,desgn):
        self.salary = salary
        self.desgn = desgn

# Deriving a child class from Two parent class
class EmpDetails(EmpInfo,OtherDtls):
    def __init__(self,name,eid,salary,desgn,expr):
        self.expr = expr
        EmpInfo.__init__(self,name,eid)
        OtherDtls.__init__(self,salary,desgn)
        print("Name: {},EmplyId: {} ,Pay: {},Post: {},Exp: {}".format(self.name,self.eid,self.salary,self.desgn,self.expr))

Sd = EmpDetails('TomHank','EMP000342',111000,'LeadArchitect',12)        
```

```python
"""
 Python Program to demonstrate decorator
"""
def somedemo(fc):
    def inner_func():
        print("Hello test")
        fc()
        print("Bye test")
    return inner_func    

@somedemo
def testf():
    print("Dead!")
    
testf()    
```

```python
"""
 Python Program to demonstrate abstract class
"""
from abc import ABC,abstractmethod
class Car(ABC):
    def drange(self):
        pass

class Tesla(Car):
    def drange(self):
        print("Range is 270km")
class NexonEvMax(Car):
    def drange(self):
        print("Range ps 150km")

y = Tesla()
y.drange()
```
